Notebook ini selaras dengan bagian time series dari `slides/04_session4_advanced.md`.

## ðŸŽ¯ Tujuan
- Parsing tanggal & membuat DateTime index
- Resampling harian, mingguan, bulanan, kuartalan
- Moving averages, rolling statistik, cumulative sum
- Growth metrics (DoD/WoW/MoM/YoY)
- Seasonality & trend decomposition

```python
import warnings
from pathlib import Path

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns

try:
    from statsmodels.tsa.seasonal import seasonal_decompose
    HAS_STATSMODELS = True
except ImportError:
    HAS_STATSMODELS = False
    print('â„¹ï¸ Install statsmodels untuk seasonal decomposition.')

warnings.filterwarnings('ignore')
plt.style.use('seaborn-v0_8-darkgrid')
sns.set_palette('viridis')
```

## ðŸ“ Load Dataset & Tentukan Kolom Tanggal

```python
DATASET_FILENAME = 'RUP-PaketPenyedia-Terumumkan-2025.parquet'
candidate_paths = [
    Path('../../../datasets/rup') / DATASET_FILENAME,
    Path('../../datasets/rup') / DATASET_FILENAME,
    Path('../datasets/rup') / DATASET_FILENAME,
    Path('datasets/rup') / DATASET_FILENAME,
]
for path in candidate_paths:
    if path.exists():
        DATA_PATH = path.resolve()
        break
else:
    raise FileNotFoundError('Dataset RUP 2025 tidak ditemukan. Pastikan folder datasets tersedia.')

df = pd.read_parquet(DATA_PATH)
print(df.shape)
```

```python
date_candidates = ['tgl_pengumuman_paket','tgl_buat_paket','tgl_awal_kontrak']
for col in date_candidates:
    if col in df.columns:
        date_col = col
        break
else:
    raise ValueError('Tidak ada kolom tanggal pada dataset.')

print('Menggunakan kolom tanggal:', date_col)
df['event_date'] = pd.to_datetime(df[date_col], errors='coerce')
df_ts = df.dropna(subset=['event_date']).copy()
df_ts = df_ts.set_index('event_date').sort_index()
print('Date range:', df_ts.index.min(), '->', df_ts.index.max())
```

## ðŸ“† Date Features

```python
df_ts['year'] = df_ts.index.year
df_ts['month'] = df_ts.index.month
df_ts['quarter'] = df_ts.index.quarter
df_ts['day_of_week'] = df_ts.index.dayofweek
df_ts['day_name'] = df_ts.index.day_name()
df_ts[['year','month','quarter','day_name']].head()
```

## ðŸ“Š Resampling (Daily/Weekly/Monthly/Quarterly)

```python
daily = df_ts.resample('D').size()
weekly = df_ts.resample('W').agg({'pagu':['count','sum','mean']})
monthly = df_ts.resample('M').agg({'pagu':['count','sum','mean']})
quarterly = df_ts.resample('Q').agg({'pagu':['count','sum']})
monthly.head()
```

```python
fig, axes = plt.subplots(3,1, figsize=(14,12))
daily.plot(ax=axes[0], title='Daily Paket Count')
monthly['pagu']['sum'].plot(ax=axes[1], title='Monthly Total Pagu (Rp)', color='coral')
quarterly['pagu']['count'].plot(kind='bar', ax=axes[2], title='Quarterly Paket Count')
plt.tight_layout()
plt.show()
```

## ðŸ“ˆ Moving Averages & EMA

```python
daily_df = daily.to_frame('count')
daily_df['SMA_7'] = daily_df['count'].rolling(7).mean()
daily_df['SMA_30'] = daily_df['count'].rolling(30).mean()
daily_df['EMA_14'] = daily_df['count'].ewm(span=14).mean()
daily_df.plot(figsize=(14,6), title='Daily Count + Moving Averages')
plt.show()
```

## ðŸ” Rolling Statistik (Â±1 Std Dev)

```python
window = 14
rolling_stats = pd.DataFrame({
    'count': daily_df['count'],
    'mean': daily_df['count'].rolling(window).mean(),
    'std': daily_df['count'].rolling(window).std()
})
fig, ax = plt.subplots(figsize=(14,6))
ax.plot(rolling_stats['count'], alpha=0.4, label='Actual')
ax.plot(rolling_stats['mean'], linewidth=2, label=f'{window}-day MA')
ax.fill_between(rolling_stats.index,
                rolling_stats['mean'] - rolling_stats['std'],
                rolling_stats['mean'] + rolling_stats['std'],
                color='orange', alpha=0.2, label='Â±1 Std Dev')
ax.legend()
plt.show()
```

## ðŸ“‰ Growth Metrics (DoD/WoW/MoM/YoY)

```python
monthly_counts = df_ts['pagu'].resample('M').count().rename('count')
monthly_df = monthly_counts.to_frame()
monthly_df['MoM_pct'] = monthly_df['count'].pct_change()*100
monthly_df['YoY_pct'] = monthly_df['count'].pct_change(12)*100
monthly_df.tail(12)
```

## ðŸ“ˆ Cumulative Sum / Running Total

```python
monthly_df['running_total'] = monthly_df['count'].cumsum()
fig, axes = plt.subplots(2,1, figsize=(14,9), sharex=True)
monthly_df['count'].plot(kind='bar', ax=axes[0], title='Monthly Paket Count')
monthly_df['running_total'].plot(ax=axes[1], title='Running Total Paket', color='green')
plt.tight_layout()
plt.show()
```

## ðŸ” Seasonality (Bulan & Hari)

```python
monthly_pattern = df_ts.groupby(df_ts.index.month).size()
ax = monthly_pattern.plot(kind='bar', figsize=(10,5), color='steelblue')
ax.set_xticklabels(['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec'], rotation=45)
ax.set_title('Seasonality: Monthly Paket Count')
plt.show()
```

```python
dow_pattern = df_ts.groupby(df_ts.index.day_name()).size().reindex(
    ['Monday','Tuesday','Wednesday','Thursday','Friday','Saturday','Sunday'])
dow_pattern.plot(kind='bar', figsize=(10,4), color='purple', title='Seasonality: Day of Week')
plt.xticks(rotation=45)
plt.show()
```

## âš™ï¸ Seasonal Decomposition

```python
if HAS_STATSMODELS and len(monthly_counts) >= 24:
    decomposition = seasonal_decompose(monthly_counts, model='additive', period=12)
    decomposition.plot()
    plt.suptitle('Seasonal Decomposition (Monthly Count)')
    plt.show()
else:
    print('Perlu statsmodels & â‰¥24 bulan data untuk decomposition.')
```

## ðŸš¨ Anomaly Detection (mean+3Ïƒ)

```python
thresh = daily_df['count'].mean() + 3*daily_df['count'].std()
anomalies = daily_df[daily_df['count'] > thresh]
print('Anomaly days:', len(anomalies))
plt.figure(figsize=(12,4))
plt.plot(daily_df['count'], label='Daily count')
plt.scatter(anomalies.index, anomalies['count'], color='red', label='Anomaly')
plt.legend()
plt.show()
```

## ðŸ“‹ Summary

```python
print('='*70)
print('TIME SERIES SUMMARY')
print('='*70)
print(f'Date range: {df_ts.index.min()} -> {df_ts.index.max()}')
print(f'Average daily count: {daily.mean():.2f}')
print(f'Month with highest count: {monthly_pattern.idxmax()}')
print(f'Seasonality: weekends count proportion {dow_pattern.iloc[-2:].sum()/dow_pattern.sum()*100:.2f}%')
print(f'Potential anomalies flagged: {len(anomalies)}')
```
